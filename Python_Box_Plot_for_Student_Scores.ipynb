```
import matplotlib.pyplot as plt
import numpy as np

# --- Data ---
# The dataset of 25 student test scores from the assignment.
student_scores = [72, 88, 94, 119, 85, 91, 77, 84, 75, 79, 83, 80, 87, 70, 76, 82, 93, 78, 89, 95, 86, 90, 73, 92, 81]

# --- Create the Box Plot ---

# Create a figure and axis object. This is the standard way to start a plot in matplotlib.
fig, ax = plt.subplots(figsize=(8, 6))

# Generate the box plot.
# The `ax.boxplot()` function automatically calculates the quartiles, median,
# and whiskers, and it identifies and plots any outliers as individual points.
# The `vert=False` argument makes the box plot horizontal for better readability of the y-axis label.
box = ax.boxplot(student_scores, vert=False, patch_artist=True,
                   flierprops={'marker': 'o', 'markerfacecolor': 'red', 'markersize': 8, 'linestyle': 'none'})

# --- Customize the Plot's Appearance ---

# Set the title and labels for clarity, as required by the assignment.
ax.set_title('Box Plot of Student Test Scores', fontsize=16, fontweight='bold')
ax.set_xlabel('Test Scores', fontsize=12)
ax.set_yticks([]) # Hide the y-axis ticks as we only have one data series.

# Add a grid for better readability of the score values.
ax.xaxis.grid(True, linestyle='--', which='major', color='grey', alpha=0.5)

# Set a background color for the plot area.
ax.set_facecolor('#f0f0f0')

# Style the box plot elements
plt.setp(box['boxes'], color='darkblue', facecolor='lightblue')
plt.setp(box['whiskers'], color='darkblue', linestyle='-')
plt.setp(box['caps'], color='darkblue')
plt.setp(box['medians'], color='navy', linewidth=2)


# --- Add Annotations for Key Statistical Points ---
# This helps to explicitly show the values on the plot.

# Calculate statistics using numpy
q1 = np.percentile(student_scores, 25)
median = np.median(student_scores)
q3 = np.percentile(student_scores, 75)
iqr = q3 - q1
upper_whisker = q3 + 1.5 * iqr
outliers = [score for score in student_scores if score > upper_whisker]

# Annotate the outlier
if outliers:
    for outlier in outliers:
        ax.text(outlier, 1.05, f'Outlier: {outlier}', verticalalignment='bottom', horizontalalignment='center', color='red', fontweight='bold')

# Annotate the median and quartiles
ax.text(median, 1.15, f'Median: {median}', verticalalignment='center', horizontalalignment='center', fontsize=10, color='navy')
ax.text(q1, 0.85, f'Q1: {q1}', verticalalignment='center', horizontalalignment='center', fontsize=10, color='darkblue')
ax.text(q3, 0.85, f'Q3: {q3}', verticalalignment='center', horizontalalignment='center', fontsize=10, color='darkblue')


# --- Display the Plot ---

# Adjust layout to prevent labels from being cut off.
plt.tight_layout()

# Show the final plot.
plt.show()
```
